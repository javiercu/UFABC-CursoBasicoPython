Existem duas formas para imprimir uma cadeia de caracteres (sequência de 1 ou mais caracteres).

A primeira seria a declaração de uma variável cujo valor atribuido seja a cadeia a ser impressa, e enseguida digitar no interpretador o nome da variável que guarda a cadeia.
Exemplo:

<code> frase = "Hello World"</code><br>
<code> frase </code>



A outra forma seria usando a função <code>print</code> que é uma função nativa do Python. Basta colocar o texto dentro dos parênteses. (Nota: Em versões anteriores de python, é possivel evitar o uso dos parênteses)

<code>print("Hello World!")</code>


Caso deseja-se imprimir o valor de uma variável, é suficiente com usar a função de saida <code>print</code> com o nome da variável dentro dos parênteses.

Por exemplo, executando o seguinte código a seguir

<code> nome = "João"</code><br>
<code> print(nome)</code>

A saída é Joao

<h2>Valores de entrada lidos pelo teclado</h2>

É possível ler do teclado as informações digitadas pelo usuário. Para isto, usa-se a função nativa <code>input</code> da seguinte forma:

<code>idade = input("Digite a sua idade: ")</code> <br>

Aparecerá:

<code>Digite a sua idade:</code>, e um quadro de texto para entrar o valor solicitado.


```python
idade = input("Digite a sua idade: ") 
```

O valor lido sempre é de tipo string (cadeia de caracteres). Caso deseja-se fazer operações sobre inteiros ou reais, deve-se converter o valor lido de entrada para os respectivos tipos (casting).
Para o exemplo da idade seria:

<code>idade = int(input("Digite sua idade: "))</code>

Dessa forma garantimos que o valor lido é um inteiro. Calculando a mitade da idade seria:

<code>print("A mitade da sua idade é ", (int)(idade/2))</code>
