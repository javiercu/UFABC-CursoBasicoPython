<h2>Valores de entrada lidos pelo teclado</h2>

 possvel ler do teclado as informaes digitadas pelo usurio. Para isto, usa-se a funo nativa <code>input</code> da seguinte forma:

<code>idade = input("Digite a sua idade: ")</code> <br>

Aparecer:

<code>Digite a sua idade:</code>, e um quadro de texto para entrar o valor solicitado.


```python
idade = input("Digite a sua idade: ") 
```

O valor lido sempre  de tipo string (cadeia de caracteres). Caso deseja-se fazer operaes sobre inteiros ou reais, deve-se converter o valor lido de entrada para os respectivos tipos (casting).
Para o exemplo da idade seria:

<code>idade = int(input("Digite sua idade: "))</code>

Dessa forma garantimos que o valor lido  um inteiro. Calculando a mitade da idade seria:

<code>print("A mitade da sua idade  ", (int)(idade/2))</code>
